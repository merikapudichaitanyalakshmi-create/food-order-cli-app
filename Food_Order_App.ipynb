**Defining Global-level Variables**

```python
# -------------------------------
# CAPO Vibes - Food Ordering CLI
# -------------------------------

RESTAURANT_NAME = "CAPO Vibes"
SALES_TAX_RATE = 0.07

menu = {
    'sku1': {'Name': 'Hamburger', 'Price': 6.51},
    'sku2': {'Name': 'Cheeseburger', 'Price': 7.75},
    'sku3': {'Name': 'Milkshake', 'Price': 5.99},
    'sku4': {'Name': 'Fries', 'Price': 2.39},
    'sku5': {'Name': 'Sub', 'Price': 5.87},
    'sku6': {'Name': 'Ice Cream', 'Price': 1.55},
    'sku7': {'Name': 'Fountain Drink', 'Price': 3.45},
    'sku8': {'Name': 'Cookie', 'Price': 3.15},
    'sku9': {'Name': 'Brownie', 'Price': 2.46},
    'sku10': {'Name': 'Sauce', 'Price': 0.75}
}

app_actions = {
    "1": "Add a new menu item to cart",
    "2": "Remove an item from the cart",
    "3": "Modify cart item quantity",
    "4": "View the cart",
    "5": "Checkout",
    "6": "Exit"
}

cart = {}
```

**Displaying the Menu**

```python
def display_menu():
    print("\n🍔 MENU")
    print("--------------------")
    for sku in menu:
        parsed_sku = sku[3:]
        item = menu[sku]['Name']
        price = menu[sku]['Price']
        print(f"({parsed_sku}) {item} - ${price}")
    print()
```

**Add Menu Items to the Cart**

```python
def add_to_cart(sku, quantity=1):
    if sku in menu:
        cart[sku] = cart.get(sku, 0) + quantity
        print(f"✅ Added {quantity} x {menu[sku]['Name']}")
    else:
        print("❌ Invalid menu item")
```

**Removing Items from the Cart**

```python
def remove_from_cart(sku):
    if sku in cart:
        del cart[sku]
        print("✅ Item removed from cart")
    else:
        print("❌ Item not in cart")
```

**Modifying Item Quantities in the Cart**

```python
def modify_cart(sku, quantity):
    if sku in cart:
        if quantity > 0:
            cart[sku] = quantity
            print("✅ Cart updated")
        else:
            remove_from_cart(sku)
    else:
        print("❌ Item not in cart")
```

**Viewing Cart Contents**

```python
def view_cart():
    print("\n🛒 Cart")
    print("--------------------")

    if not cart:
        print("Cart is empty")
        return

    subtotal = 0

    for sku, quantity in cart.items():
        item = menu[sku]['Name']
        price = menu[sku]['Price']
        subtotal += price * quantity
        print(f"{quantity} x {item}")

    tax = subtotal * SALES_TAX_RATE
    total = subtotal + tax

    print("--------------------")
    print(f"Subtotal: ${round(subtotal, 2)}")
    print(f"Tax: ${round(tax, 2)}")
    print(f"Total: ${round(total, 2)}\n")
```

**Checking Out**

```python
def checkout():
    print("\n💳 Checkout")
    view_cart()
    print(f"Thank you for ordering at {RESTAURANT_NAME}!\n")
```

**Prompting for User Input**

```python
def get_sku_and_quantity():
    try:
        sku = input("Enter item number: ")
        sku = "sku" + sku

        quantity = input("Enter quantity: ")

        if not quantity.isdigit():
            quantity = 1

        return sku, int(quantity)

    except:
        print("Invalid input")
        return None, 1
```

**Building the Ordering Loop**

```python
def order_loop():
    print(f"\nWelcome to {RESTAURANT_NAME}!")

    ordering = True

    while ordering:
        print("\n📋 Actions")
        print("--------------------")

        for number, action in app_actions.items():
            print(f"({number}) {action}")

        choice = input("Choose option: ")

        if choice == "1":
            display_menu()
            sku, quantity = get_sku_and_quantity()
            add_to_cart(sku, quantity)

        elif choice == "2":
            display_menu()
            sku = "sku" + input("Enter item number: ")
            remove_from_cart(sku)

        elif choice == "3":
            display_menu()
            sku, quantity = get_sku_and_quantity()
            modify_cart(sku, quantity)

        elif choice == "4":
            view_cart()

        elif choice == "5":
            checkout()
            ordering = False

        elif choice == "6":
            print("👋 See you!")
            ordering = False

        else:
            print("❌ Invalid option")
```

**Testing the Application**

```python
if __name__ == "__main__":
    order_loop()
```
